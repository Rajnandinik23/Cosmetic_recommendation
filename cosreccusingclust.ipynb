```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/jjone36/Cosmetic/master/data/cosmetic_p.csv')
 # Replace 'cosmetic_data.csv' with your dataset file

```

```python
data
```

```python
features = data['ingredients']
labels = data['Combination']

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Step 4: Transform text data into numerical features using TF-IDF vectorization
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)

# Step 5: Train a Nearest Neighbors model
k = 5  # Number of neighbors to consider
model = NearestNeighbors(n_neighbors=k)
model.fit(X_train)

# Step 6: Recommend cosmetics based on user input
user_input = "Phenylalanine, Proline, Serine, Threonine,Niacinamide"  # Replace with user's desired ingredients
user_input = vectorizer.transform([user_input])

distances, indices = model.kneighbors(user_input)

# Step 7: Display recommended cosmetics
print("Recommended cosmetics:")
for idx in indices.flatten():
  if(data['Combination'].iloc[idx]==1):
    print(data['Label'].iloc[idx],'|',data['brand'].iloc[idx],'|',data['name'].iloc[idx])
```
